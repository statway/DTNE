```python
import pickle
import time
with open('./mnist_X.dat', 'rb') as infile:
    X = pickle.load(infile)
with open('./mnist_color.dat', 'rb') as infile:
    color = pickle.load(infile)
```

```python
import matplotlib.pyplot as plt
```

```python
from dtne import *
```

```python
start_time = time.time()
embedding = DTNE(k_neighbors =10, l=10, kd='t')
Y = embedding.fit_transform(X)
end_time = time.time()
print('The code runs {:.0f}s.'.format(end_time - start_time))
plt.scatter(Y[:,0], Y[:,1], c=color,s=0.01,cmap='jet')
```

```python
start_time = time.time()
embedding = DTNE(k_neighbors =10, l=10, kd='log')
Y = embedding.fit_transform(X)
end_time = time.time()
print('The code runs {:.0f}s.'.format(end_time - start_time))
plt.scatter(Y[:,0], Y[:,1], c=color,s=0.01,cmap='jet')
```

```python
start_time = time.time()
embedding = DTNE(k_neighbors =10,l1=20)
Y = embedding.fit_transform(X)
end_time = time.time()
print('The code runs {:.0f}s.'.format(end_time - start_time))
plt.scatter(Y[:,0], Y[:,1], c=color,s=0.01,cmap='jet')
```
