```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("runtime.csv",index_col=0,header=[0]) 
data
```

```python
method_names = ["DTNE","PHATE","UMAP","tSNE"]
```

```python
fig, ax = plt.subplots()

markers = ["s","d","^",'o',]

color2=[
    '#1f77b4', '#aec7e8', '#ff7f0e', '#ffbb78', '#2ca02c', '#98df8a',
    '#d62728', '#ff9896', '#9467bd', '#c5b0d5', '#8c564b', '#c49c94',
    '#e377c2', '#f7b6d2', '#7f7f7f', '#c7c7c7', '#bcbd22', '#dbdb8d',
    '#17becf', '#9edae5']  

for i in range(4):
    line, = ax.plot(data.iloc[0,:].values, data.iloc[i+1,:].values, marker = markers[i],color= color2[i+0]) # # color=line.get_color()
    ax.scatter(data.iloc[0,:].values, data.iloc[i+1,:].values, marker = markers[i], color= color2[i+0])
    ax.plot([], [], marker = markers[i], label = method_names[i], color=color2[i+0])
plt.legend(loc='best') 

plt.xlabel('Number of samples for input data', fontsize=12)
plt.ylabel('Average run time', fontsize= 12)

sns.despine(top=True,right=True,left=False,bottom=False,)

plt.savefig('./sc_runtime.png', format='png', dpi=1000)
plt.savefig('./sc_runtime.pdf', format='pdf', dpi=1000)
```
